```python
def prepare_sentiment_data(company_name, filename):
  from google.colab import drive

  drive.mount('/content/gdrive')

  news_data = pd.read_excel(r"/content/gdrive/My Drive/Dataset/"+filename,sheet_name=company_name)
  news_data['date']=pd.to_datetime(news_data['date']).dt.date

  news_data.dropna(inplace=True)
  news_data['sentiment'] = news_data['sentiment'].apply(ast.literal_eval)
  news_data['polarity score'] = news_data['sentiment'].apply(lambda x: x.get('polarity'))

  polarity = news_data.groupby('date')['polarity score'].mean()
  polarity = pd.DataFrame(polarity)
  polarity = polarity.rename(columns={'polarity score':company_name+'_polarity'})

  return polarity
```

```python
data["Future_Return"] = (data["Close"].shift(-1) - data["Close"]) / data["Close"]  # Next day return

# Momentum crossover strategy
data["Alpha_1"] = data["Momentum_3"] - data["Momentum_10"]

# Bollinger Bands normalization
data["Alpha_2"] = (data["Close"] - data["BB_Lower"]) / ((data["BB_Upper"] - data["BB_Lower"]) + 1e-6)

# MACD crossover signal
data["Alpha_3"] = data["MACD"] - data["MACD_Signal"]

# RSI-based mean reversion (normalized)
data["Alpha_4"] = (data["RSI_14"] - 50) / 100.0

data["SMA_20_Volume"] = data.ta.sma(length=20, close=data["Volume"])
# Volume and trend confirmation
data["Alpha_5"] = (data["Volume"] / (data["SMA_20_Volume"] + 1e-6)) * (data["Close"] - data["SMA_5"])
```

```python
data["Future_Return"] = (data["Close"].shift(-1) - data["Close"]) / data["Close"]  # Next day return

# Alpha 1: Average of short-term momentum + average of Japanese automakers' sentiment
data["Alpha_1"] = ((data["Momentum_3"] + data["Momentum_10"]) / 2) + (
    (data["Toyota_polarity"] + data["Nissan_polarity"] + data["Honda_polarity"]) / 3
)

# Alpha 2: Oversold RSI condition + positive Toyota sentiment
data["Alpha_2"] = data["RSI_14"] * (
    ((data["RSI_14"] < 30).astype(int)) * ((data["Toyota_polarity"] > 0).astype(int))
)

# Alpha 3: Bullish SMA crossover + positive Toyota sentiment
data["Alpha_3"] = ((data["SMA_5"] > data["SMA_20"]) & (data["Toyota_polarity"] > 0)).astype(int)

# Alpha 4: Breakout potential + positive Apple and Amazon sentiment
data["Alpha_4"] = ((data["Close"] - data["BB_Lower"]) / (data["BB_Upper"] - data["BB_Lower"])) * (
    ((data["Apple_polarity"] > 0) & (data["Amazon_polarity"] > 0)).astype(int)
)

# Alpha 5: Bullish OBV and price action + positive EV sentiment (BYD and TSLA)
data["Alpha_5"] = (
    ((data["OBV"] > data["OBV"].shift(1)) & (data["Close"] > data["Close"].shift(1)) &
     (data["BYD_polarity"] > 0) & (data["TSLA_polarity"] > 0)).astype(int)
)
```

```python
# 📌 Function to compute Information Coefficient (IC)
def compute_ic(alpha_values, future_returns):
    ic = stats.spearmanr(alpha_values, future_returns)[0]  # Rank correlation
    return ic

# 📌 Compute IC (how well Alpha predicts future returns)
for i in range(1, 6):
    ic_value = compute_ic(data[f"Alpha_{i}"], data["Future_Return"])
    print(f"Information Coefficient (IC) for Alpha_{i}: {ic_value:.4f}")
```
