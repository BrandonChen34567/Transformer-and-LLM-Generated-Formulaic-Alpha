```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
import numpy as np
from groq import Groq
import pandas_ta as ta
```

```python
def prepare_sentiment_data(company_name, filename):
  from google.colab import drive

  drive.mount('/content/gdrive')

  news_data = pd.read_excel(r"/content/gdrive/My Drive/Dataset/"+filename,sheet_name=company_name)
  news_data['date']=pd.to_datetime(news_data['date']).dt.date

  news_data.dropna(inplace=True)
  news_data['sentiment'] = news_data['sentiment'].apply(ast.literal_eval)
  news_data['polarity score'] = news_data['sentiment'].apply(lambda x: x.get('polarity'))

  polarity = news_data.groupby('date')['polarity score'].mean()
  polarity = pd.DataFrame(polarity)
  polarity = polarity.rename(columns={'polarity score':company_name+'_polarity'})

  return polarity
```

```python
import featuretools as ft
```

```python
data["index"] = range(len(data))  # create a unique row index for Featuretools

# Step 1: Create an EntitySet
es = ft.EntitySet(id="stock_data")

# Step 2: Add your DataFrame to the EntitySet
es = es.add_dataframe(
    dataframe_name="stock",
    dataframe=data,
    index="index",
    time_index=data.index.name  # this uses 'date' as the time index
)

# Step 3: Define transformation primitives
trans_primitives = [
    # Arithmetic
    "add_numeric", "subtract_numeric", "multiply_numeric", "divide_numeric",

    # Absolute/Ratio
    "absolute", "negate", "percentile", "divide_by_feature",

    # Cumulative
    "cum_sum", "cum_mean", "cum_max", "cum_min", "cum_count",

    # Polynomial
    "square_root",  "modulo_by_feature", "diff"
]


# Step 4: Apply DFS
feature_matrix, feature_defs = ft.dfs(
    entityset=es,
    target_dataframe_name="stock",
    trans_primitives=trans_primitives,
    max_depth=2  # or 2 for more complexity
)

```

```python
feature_matrix = feature_matrix.fillna(0)
feature_matrix
```

```python
# feature 1: Close + Honda_polarity
data['feature 1'] = data['Close'] + data['Honda_polarity']

# feature 2: BMW_polarity + Close
data['feature 2'] = data['BMW_polarity'] + data['Close']

# feature 3: BYD_polarity + Close
data['feature 3'] = data['BYD_polarity'] + data['Close']

# feature 4: Close - Ford_polarity
data['feature 4'] = data['Close'] - data['Ford_polarity']

# feature 5: Close - GM_polarity
data['feature 5'] = data['Close'] - data['GM_polarity']
```

```python
top5_features = ['feature 1', 'feature 2', 'feature 3', 'feature 4', 'feature 5']
```

```python
X = top5_features + ['Close']
```

```python
stock_data_train = data[X].iloc[:round(len(data)*0.7)]
stock_data_test = data[X][round(len(data)*0.7):]
```

```python
window = 5
predict_forward = 1

# =========================
# Feature Scaling (TRAIN)
# =========================
scaler_x = MinMaxScaler(feature_range=(0, 1))
train_data = scaler_x.fit_transform(stock_data_train)

scaler_y = MinMaxScaler(feature_range=(0, 1))
train_target = scaler_y.fit_transform(
    stock_data_train['Close'].values.reshape(-1, 1)
)

# =========================
# Build Training Sequences
# =========================
X_train, y_train = [], []

for i in range(len(train_data) - window - predict_forward + 1):
    X_train.append(train_data[i:i + window])
    y_train.append(train_target[i + window:i + window + predict_forward])

X_train = np.array(X_train)
y_train = np.array(y_train).reshape(-1, predict_forward)

# =========================
# Feature Scaling (TEST)
# =========================
test_data = scaler_x.transform(stock_data_test)
test_target = scaler_y.transform(
    stock_data_test['Close'].values.reshape(-1, 1)
)

# =========================
# Build Testing Sequences
# =========================
X_test, y_test = [], []

for i in range(len(test_data) - window - predict_forward + 1):
    X_test.append(test_data[i:i + window])
    y_test.append(test_target[i + window:i + window + predict_forward])

X_test = np.array(X_test)
y_test = np.array(y_test).reshape(-1, predict_forward)
```
